```python
#importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\molwa\OneDrive\Documents\Machine Learnig Projects\datasets\president_heights.csv")
data
```

```python
#check the shape of the data
data.shape
```

```python
#check if there are null values in the data
data.isnull().sum()
```

```python
#check the stastics of the data
data.describe().T
```

```python
#convert the height column to a numpy array
height = np.array(data['height(cm)'])
print(height)
```

```python
#compute the stastics of the height and see if it accurate to what we got above
print('The mean of the height is: {} \nThe standard deviation of the height is: {} \nThe minimum height is: {} \nThe maximum height is: {}'.format(height.mean(), height.std(), height.min(), height.max()))
```

```python
#Find the percentile of the data
print('The 25th percentile: ', np.percentile(height, 25))
print('The median: ', np.median(height))
print('The 75th percentile: ', np.percentile(height, 75))
```

```python
plt.hist(height)
plt.title('The Distribution Heights of USA presidents')
plt.xlabel('The height of the presidents in cm')
plt.ylabel('Count')
plt.show()
```
